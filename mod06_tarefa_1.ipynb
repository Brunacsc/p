#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
# importar bibliotecas e criar df

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sinasc_df = pd.read_csv('/content/b1abb33c4f.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Criar um gráfico de barras para as seguintes variáveis

import matplotlib.pyplot as plt

# Criar o plot e eixos
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

# Para a variável 'APGAR1'
axes[0, 0].bar(sinasc_df['APGAR1'].value_counts().index, sinasc_df['APGAR1'].value_counts().values)
axes[0, 0].set_title('APGAR1')

# Para a variável 'APGAR5'
axes[0, 1].bar(sinasc_df['APGAR5'].value_counts().index, sinasc_df['APGAR5'].value_counts().values)
axes[0, 1].set_title('APGAR5')

# Para a variável 'SEXO'
axes[1, 0].bar(sinasc_df['SEXO'].value_counts().index, sinasc_df['SEXO'].value_counts().values)
axes[1, 0].set_title('SEXO')

# Para a variável 'GESTACAO'
axes[1, 1].bar(sinasc_df['GESTACAO'].value_counts().index, sinasc_df['GESTACAO'].value_counts().values)
axes[1, 1].set_title('GESTACAO')

# Para a variável 'ESTCIVMAE'
axes[2, 0].bar(sinasc_df['ESTCIVMAE'].value_counts().index, sinasc_df['ESTCIVMAE'].value_counts().values)
axes[2, 0].set_title('ESTCIVMAE')

# Ajuste de exibição
plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
#Gráfico de barras mostrando o APGAR5 médio por CONSULTAS

# Calcular a média
apgar5_por_consultas = sinasc_df.groupby('CONSULTAS')['APGAR5'].mean()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(apgar5_por_consultas.index, apgar5_por_consultas.values)
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')
plt.title('APGAR5 Médio por Número de Consultas')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Criar o plot e eixos
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Para a variável 'SEXO'
axes[0, 0].hist(sinasc_df['SEXO'].astype(str), bins=sinasc_df['SEXO'].nunique()) # utilizado o .astype(str), pois estava dando erro
axes[0, 0].set_title('Frequência de SEXO')

# Para a variável 'GESTACAO'
axes[0, 1].hist(sinasc_df['GESTACAO'].astype(str), bins=sinasc_df['GESTACAO'].nunique())
axes[0, 1].set_title('Frequência de GESTACAO')

# Para a variável 'ESTCIVMAE'
axes[1, 0].hist(sinasc_df['ESTCIVMAE'].astype(str), bins=sinasc_df['ESTCIVMAE'].nunique())
axes[1, 0].set_title('Frequência de ESTCIVMAE')

# Para a variável 'CONSULTAS'
axes[1, 1].hist(sinasc_df['CONSULTAS'], bins=sinasc_df['CONSULTAS'].nunique())
axes[1, 1].set_title('Frequência de CONSULTAS')

# Ajuste de exibição
plt.tight_layout()
plt.show()
```
