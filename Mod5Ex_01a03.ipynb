````python
import pandas as pd
import requests

# 1) seu código aqui

# Definir o caminho do arquivo
caminho_arquivo = '/content/SINASC_RO_2019.csv'

# Ler o arquivo e definir uma variável para ele
sinasc = pd.read_csv(caminho_arquivo)

# Exibi a quantidade de linhas do arquivo
print('A quantidade de linhas do arquivo é:', sinasc.shape)

# Exibe a quantidade de linhas duplicadas
duplicadas = sinasc.drop_duplicates().shape

# Não há duplicados
print('Não há linhas duplicadas, pois temos o mesmo valor da quantidade de linhas do arquivo:', duplicadas )
````

````python
# 2) seu código aqui

# Conta o número de valores ausentes por variáveis
valores_ausentes = sinasc.isnull().sum()

# Exibe o resultado
print('Número de valores ausentes por variáveis \t', valores_ausentes)
````

````python
# 3) seu código aqui

# Criar um df para as variáveis selecionadas
valores_importantes = pd.DataFrame(['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO',
 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5'] )

# Contar os números de valores ausentes nas variáveis selecionadas
valores_ausentes_selecionados = valores_importantes.isnull().sum()

# Exibe o resultado
print('O número de valores ausentes nas variáveis selecionadas é', valores_ausentes_selecionados)
````

````python
# 5) seu código aqui

# Criado um novo Df para as duas variáveis selecionadas
est_consulta = pd.DataFrame(['ESTCIVMAE', 'CONSULTAS'])

# Substituir o número '9' por NaN
est_consulta.replace(9, pd.NA, inplace=True)

# Exibe o número de dados preenchidos com o número '9' nas varíaveis selecionadas
print('O número de dados ausentes é', est_consulta.isnull().sum())
````

````python
# 6) Seu código aqui

Qtd_filho_vivo = sinasc['QTDFILVIVO']

# Exibe a quantidade de dados faltantes dentro da variável 'QTDDILVIVO'
print('A quantidade de NaN é:', Qtd_filho_vivo.isnull().sum())

# Substituir o NaN por '0'
Qtd_filho_vivo.fillna(0, inplace=True)

# Exibe a quantidade de dados faltantes após a troca
print('A quantidade de NaN após a alteração é:', Qtd_filho_vivo.isnull().sum())
````

````python
# 7) seu código aqui

# Irei substituir o NaN por '0' para que o Df fique completamente preechida #
sinasc.fillna(0, inplace=True)

# Exibe a quantidade de dados faltantes em todo o Df após a troca
print('A quantidade de NaN após a alteração é:', sinasc.isnull().sum())
````

````python
# 8) seu código aqui

# Criar uma nova coluna 'CategoriaAPGAR' com base nas faixas fornecidas
sinasc['CategoriaAPGAR'] = pd.cut(sinasc['APGAR5'], bins=[-1, 3, 5, 7, 10], labels=['asfixia severa', 'asfixia moderada', 'asfixia leve', 'normal'])

# Calcular e imprimir as frequências da nova categoria
frequencias = sinasc['CategoriaAPGAR'].value_counts()

# Exibe a frequência para a nova categoria
print(frequencias)
````

````python
# 9) seu código aqui

# Renomear as colunas para o formato snake case
sinasc_atualizado = sinasc.rename(columns=lambda x: x.lower().replace(' ', '_'),inplace=False)

# Exibir o DataFrame após a renomeação
sinasc_atualizado
````
