```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
#Abrir o arquivo base
sample_data = pd.read_csv('/content/b1abb33c4f.csv')
sample_data.head()
```

```python
#Definir o Boxplot

plt.figure(figsize=(8, 6))
sns.boxplot(x=sample_data['IDADEMAE'])
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
#Definir o boxxplot do Peso do Bebê
plt.figure(figsize=(8, 6))
sns.boxplot(x=sample_data['PESO'])
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
#Histograma da idade da mãe

plt.figure(figsize=(8, 6))
sns.histplot(sample_data['IDADEMAE'], kde= 'TRUE')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
#Histograma para o Peso do Bebê

plt.figure(figsize=(8, 6))
sns.histplot(sample_data['PESO'], kde= 'TRUE')
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso do Bebê')
plt.ylabel('Frequência')
plt.show()
```
